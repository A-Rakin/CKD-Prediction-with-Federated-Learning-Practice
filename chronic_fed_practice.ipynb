```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import StratifiedKFold

# Step 1: Load dataset
df = pd.read_csv("Chronic_Kidney_Dsease_data.csv")
df = df.drop(columns=["PatientID", "DoctorInCharge"])

# Step 2: Split into clients
clients = np.array_split(df, 3)

n_splits = 5
global_model_coefs = []
global_model_intercepts = []

print("=== 5-Fold CV Federated Learning with SMOTE + Scaling ===")

for fold in range(n_splits):
    print(f"\n----- Fold {fold+1} -----")

    local_models = []
    scalers = []

    # Train local models per client
    for i, client_data in enumerate(clients):
        X_client = client_data.drop("Diagnosis", axis=1)
        y_client = client_data["Diagnosis"]

        # Stratified split for this fold
        skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
        for j, (train_idx, test_idx) in enumerate(skf.split(X_client, y_client)):
            if j != fold:
                continue  # pick only current fold

            X_train_client, X_test_client = X_client.iloc[train_idx], X_client.iloc[test_idx]
            y_train_client, y_test_client = y_client.iloc[train_idx], y_client.iloc[test_idx]

            # SMOTE on training data only
            smote = SMOTE(random_state=42)
            X_train_res, y_train_res = smote.fit_resample(X_train_client, y_train_client)

            # Scaling
            scaler = StandardScaler()
            X_train_scaled = scaler.fit_transform(X_train_res)
            X_test_scaled = scaler.transform(X_test_client)

            # Train local logistic regression
            model = LogisticRegression(max_iter=2000, class_weight='balanced')
            model.fit(X_train_scaled, y_train_res)

            local_models.append(model)
            scalers.append(scaler)

            # Optional: local test evaluation
            y_pred_test = model.predict(X_test_scaled)
            print(f"Client {i+1} Fold {fold+1} Test Accuracy: {accuracy_score(y_test_client, y_pred_test):.3f}")

    # Federated averaging (global model for this fold)
    global_coef = np.mean([m.coef_ for m in local_models], axis=0)
    global_intercept = np.mean([m.intercept_ for m in local_models], axis=0)

    global_model_coefs.append(global_coef)
    global_model_intercepts.append(global_intercept)

# Step 2: Average global model over folds
final_global_coef = np.mean(global_model_coefs, axis=0)
final_global_intercept = np.mean(global_model_intercepts, axis=0)

global_model = LogisticRegression(max_iter=2000, class_weight='balanced')
global_model.coef_ = final_global_coef
global_model.intercept_ = final_global_intercept
global_model.classes_ = np.array([0,1])

```

```python
# Step 1: Combine all clients' test data across folds
X_test_all = []
y_test_all = []

for i, client_data in enumerate(clients):
    X_client = client_data.drop("Diagnosis", axis=1)
    y_client = client_data["Diagnosis"]

    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    for train_idx, test_idx in skf.split(X_client, y_client):
        X_test_fold = X_client.iloc[test_idx]
        y_test_fold = y_client.iloc[test_idx]

        # Scale using client scaler
        X_test_scaled = scalers[i].transform(X_test_fold)

        X_test_all.append(X_test_scaled)
        y_test_all.append(y_test_fold.values)

X_test_all = np.vstack(X_test_all)
y_test_all = np.concatenate(y_test_all)

# Step 2: Global model prediction
y_pred_global = global_model.predict(X_test_all)

# Step 3: Evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("\n===== Global Federated Model Performance after 5-Fold CV =====")
print("Accuracy:", accuracy_score(y_test_all, y_pred_global))
print("Confusion Matrix:\n", confusion_matrix(y_test_all, y_pred_global))
print("Classification Report:\n", classification_report(y_test_all, y_pred_global))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, RocCurveDisplay, ConfusionMatrixDisplay

# Confusion Matrix
cm = confusion_matrix(y_test_all, y_pred_global)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
disp.plot(cmap=plt.cm.Blues)
plt.title("Global Federated Model Confusion Matrix")
plt.show()

# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test_all, global_model.predict_proba(X_test_all)[:,1])
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0,1], [0,1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Global Federated Model ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
